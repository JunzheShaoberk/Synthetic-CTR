```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
# Load the data

df = pd.read_csv('CTR_data_v1_2023-March_25_without_name_fct_privacy.csv')
df.head()
```

```python
# select ALL columns under the following categories
# columns do not start with inj
# and columns start with inj_loc
# and columns start with inj_fractureclosed.

df1 = df.loc[:, ~df.columns.str.startswith('inj')]
df2 = df.loc[:, df.columns.str.startswith('inj_loc')]
df3= df.loc[:, df.columns.str.startswith('inj_fractureclosed')]
df_fracture = pd.concat([df1, df2, df3], axis=1)
df_fracture.head()

# save df_fracture to a csv file
df_fracture.to_csv('df_fracture.csv', index=False)
```
