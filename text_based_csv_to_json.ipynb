```python
import pandas as pd
import json
import random

def csv_to_json(csv_file, json_file, M=5):
    # Read the CSV file
    df = pd.read_csv(csv_file)
    
    # Convert each row to a sentence
    sentences = []
    for _, row in df.iterrows():
        sentence = ", ".join([f"{col} is {row[col]}" for col in df.columns])
        sentences.append(sentence)
    
    # Create a list to store the json entries
    json_entries = []
    
    # Split each sentence into "input" and "output" parts randomly M times
    for sentence in sentences:
        for _ in range(M):
            words = sentence.split(", ")
            random.shuffle(words)
            split_index = random.randint(1, len(words)-1)
            input_part = ", ".join(words[:split_index])
            output_part = ", ".join(words[split_index:])
            json_entries.append({
                "prompt": "Complete the sentence generated by a csv file:",
                "input": input_part,
                "output": output_part
            })
    
    # Save to JSON file
    with open(json_file, 'w') as f:
        json.dump(json_entries, f, indent=4)
```

```python
csv_file = 'df_fracture_subset.csv'
json_file = 'output_synthetic_text.json'
csv_to_json(csv_file, json_file, M=2)
```
